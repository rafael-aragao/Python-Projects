ABSTRAÇÃO

```python
class ControleRemoto:
    def __init__(self):
        self.ligado = False
        self.volume = 10 

    def ligar(self):
        self.ligado = True
        print("TV ligada!")

    def desligar(self):
        self.ligado = False
        print("TV desligada!")

    def aumentar_volume(self):
        if not self.ligado:
            print("Não dá para aumentar o volume: a TV está desligada.")
            return

        if self.volume < 20:
            self.volume += 1
            print(f"Volume: {self.volume}")
        else:
            print("Volume já está no máximo (20).")

    def diminuir_volume(self):
        if not self.ligado:
            print("Não dá para diminuir o volume: a TV está desligada.")
            return

        if self.volume > 0:
            self.volume -= 1
            print(f"Volume: {self.volume}")
        else:
            print("Volume já está no mínimo (0).")

    def status(self):
        estado = "Ligada" if self.ligado else "Desligada"
        print(f"Status: {estado} | Volume: {self.volume}")
```
