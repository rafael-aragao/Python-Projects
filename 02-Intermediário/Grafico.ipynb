```python
import matplotlib.pyplot as plt

meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
vendas = [10, 34, 20, 90, 12, 190, 23, 0, 90, 10, 23, 22]

plt.plot(meses, vendas, marker = "o")
plt.xlabel("Mês")
plt.ylabel("Vendas do Playstation 6 master chief edition")
plt.title("Vendas da Sony 2025")
plt.grid(True)
plt.show()
```
